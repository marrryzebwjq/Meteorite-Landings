```python
import pandas as pd

# Load the CSV file
file_path = '../data/Meteorite_Landings_20240924.csv'
# read only the first 25 rows
df = pd.read_csv(file_path, nrows=25)

df
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="http")

# for each line of the dataframe, get the country code
def get_country_code(row):
    location = geolocator.reverse(f"{row['reclat']}, {row['reclong']}") # take the latitude and longitude and get the location
    try:
        tmp = location.raw.get('address').get('country_code') # get the country code from the location
    except:
        tmp = None
    
    print(tmp)
    return tmp
    

df['country_code'] = df.apply(get_country_code, axis=1) # apply the function to the dataframe
df
```

```python
# Read the entire CSV file
df_full = pd.read_csv(file_path)

# Apply the get_country_code function to the entire dataframe
df_full['country_code'] = df_full.apply(get_country_code, axis=1)


# Save the dataframe to a new CSV file
df_full.to_csv('../data/Meteorite_Landings_20240924_processed.csv', index=False)
```
